```python
fig = plt.figure(figsize=(8,4))
ax1 = fig.add_subplot(1,2,1)
x = np.linspace(-5,5,21)
y1 = 3*x+1
y2 = (-1/3)*x+1
ax1.plot(x,y1)
ax1.plot(x,y2)

ax2 = fig.add_subplot(1,2,2)
ax2.plot(x,y1)
ax2.plot(x,y2)
ax2.set_aspect('equal')

plt.show()
```

#### 4. Задание (на листочке)  
Пусть задана плоскость:  

$ A \cdot x + B \cdot y + C \cdot z + D = 0 $

Напишите уравнение плоскости, параллельной данной и проходящей через начало координат.

$ A \cdot x + B \cdot y + C \cdot z = 0 $

___
Пусть задана плоскость   

$ A_1 \cdot x + B_1 \cdot y + C_1 \cdot z + D_1 = 0 $  
и прямая:  
$\frac{x-x_1}{x_2-x_1} = \frac{y-y_1}{y_2-y_1} = \frac{z-z_1}{z_2-z_1}$  
Как узнать, принадлежит прямая плоскости или нет?  

Для этого точка прямой должна принадлежать плоскости и прямая должна быть перпендикулярна вектору нормали плоскости  
Параллельность будет выполняться в случае, когда скалярное произведение векторов нормали плоскости и вектора прямой будет равно нулю.  
Или  

$A \cdot (x_2-x_1)+B \cdot (y_2-y_1)+C \cdot (z_2-z_1) = 0$

Принадлежность точки прямой плоскости определим, подставив в уравнение плоскости точку, принадлежащую прямой  

$ A_1 \cdot x_1 + B_1 \cdot y_1 + C_1 \cdot z_1 + D_1 = 0 $

#### 6. Задание (в программе)

$ y(x) = k \cdot \cos(x – a) + b $  
для некоторых (2-3 различных) значений параметров k, a, b


```python
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
k, a, b = 0.5, 0.1, -3
axes()
for i in range(3):
    k += 2*i
    a += i/2
    b += i*3
    y = k * np.cos(x - a) + b
    plt.plot(x, y)
plt.show()
```

```python
def polar2dc(r,phi):
    x = r*np.cos(phi)
    y = r*np.sin(phi)
    return x,y

polar2dc(5, 30)
```

```python
fig = plt.figure(figsize=(10,10))

r = np.arange(0, 1.5, 0.01)
theta = 2 * np.pi * r

ax1 = fig.add_subplot(221, projection='polar')
ax1.plot(theta, r)
ax1.set_rmax(2)
ax1.set_rticks([0.5, 1, 1.5, 2])
ax1.set_rlabel_position(-22.5)
ax1.grid(True)

phi = np.linspace(0, 2*np.pi, 1000)
R = 1.5
x = R*np.cos(phi)
y = R*np.sin(phi)
ax2 = fig.add_subplot(222, projection='polar')
ax2.plot(x, y)
ax2.set_rmax(2)
ax2.set_rticks([0.5, 1, 1.5, 2])
ax2.set_rlabel_position(-22.5)
ax2.grid(True)

ax3 = fig.add_subplot(223, projection='polar')
ax3.plot(np.linspace(0, 2*np.pi, 100), np.ones(100)*1.5, color='r', linestyle='-')
ax3.set_rmax(2)
ax3.set_rticks([0.5, 1, 1.5, 2])
ax3.set_rlabel_position(-22.5)
ax3.grid(True)

plt.show()
```
